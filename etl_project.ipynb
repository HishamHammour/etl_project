```python
Base = declarative_base()
class test(Base):
    #Tell SQLAlchemy what the table name is and if there's any table-specific arguments it should know about
    __tablename__ = 'test_table'
    __table_args__ = {'sqlite_autoincrement': True}
    #tell SQLAlchemy the name of column and its attributes:
    id = Column(Integer, primary_key=True, nullable=False) 
    Neighborhood = Column(String(255))
    sector_number = Column("Sector #", String(255))
    pop_2010 = Column("Population (2010)", String(255))
    total_pop_25andup = Column("Total Pop, 25 and older (2010)", String(255)) 
    less_than_highsch = Column("Edu. Attainment: Less than High School (2010)", String(255))
    highsch_grd = Column("Edu. Attainment: High School Graduate (2010)", String(255))
    assoc_prf_deg = Column("Edu. Attainment: Assoc./Prof. Degree (2010)", String(255))
    bac_deg = Column("Edu. Attainment: Bachelor's Degree (2010)", String(255))
    postgrd_deg = Column("Edu. Attainment: Postgraduate Degree (2010)", String(255))
    median_income_1999 = Column("1999 Median Income ('99 Dollars)", String(255))
    median_income_2009 = Column("2009 Median Income ('09 Dollars)", String(255))
    median_income_2011_dlrs = Column("1999 Median Income ('11 Dollars)", String(255))
    median_income_2013_dlrs = Column("2009 Med. Income ('13 Dollars)", String(255))
    est_pop_poverty_cal = Column("Est. Pop. for which Poverty Calc. (2010)", String(255))
    est_pop_under_poverty = Column("Est. Pop. Under Poverty (2010)", String(255))
    est_pop_pct_under_poverty = Column("Est. Percent Under Poverty (2010)", String(255))
    

    
```

```python
#Create the database
engine = create_engine('sqlite:///csv_test.sqlite')
conn = engine.connect()
Base.metadata.create_all(conn)
```

```python
df1.to_sql(test.__tablename__, conn, index=False, if_exists='append')
```
